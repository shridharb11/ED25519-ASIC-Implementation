```python
import struct, hashlib

pattern = [0xDEADBEEF, 0xCAFEBABE, 0x12345678, 0x87654321]
total_words = 65536  # 256 KB / 4 bytes per word

words = [pattern[i % len(pattern)] for i in range(total_words)]

with open("machine_code.hex", "w") as f:
    for w in words:
        f.write(f"{w:08x}\n")

# Compute SHA-512 over the raw big-endian byte stream
raw_bytes = b''.join(struct.pack('>I', w) for w in words)
digest = hashlib.sha512(raw_bytes).hexdigest()  # 128 hex chars = 16 x 32-bit words

# Write hash.hex as 16 lines of 8 hex chars (32-bit words, MSW first)
# $readmemh will load these into a logic[31:0] array[0:15]
with open("hash.hex", "w") as f:
    for i in range(16):
        f.write(digest[i*8 : i*8+8] + "\n")

print(f"Generated machine_code.hex with {total_words} words (256 KB).")
print(f"SHA-512: {digest}")
```
